```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
#Loading the file
df=pd.read_csv(r"C:\Users\devma\Downloads\Car_price.csv")
df.head()
```

```python
#Changing the columns of the data with the headers
headers = ["symboling", "normalized-losses", "make",  
           "fuel-type", "aspiration","num-of-doors", 
           "body-style","drive-wheels", "engine-location", 
           "wheel-base","length", "width","height", "curb-weight", 
           "engine-type","num-of-cylinders", "engine-size",  
           "fuel-system","bore","stroke", "compression-ratio", 
           "horsepower", "peak-rpm","city-mpg","highway-mpg","price"] 
  
df.columns=headers 
df.head() 
```

```python
data=df

#Finding the missing values
data.isna().any()
```

```python
#Converting the miles per gallons to Litres per 100K.m.
data['city-mpg']=235/df['city-mpg']
data.rename(columns={'city-mpg':"city-L/100km"},inplace=True)
print(data.columns)
data.dtypes
```

```python
data.price.unique()
```

```python
#removing the '?'
data=data[data.price != '?']
data.dtypes
```

```python
#Price should be int64 type but its object
#Converting the data type
data['price']=data['price'].astype(str).astype('int64')
```

```python
#Normalizing the length, width,height such that it will be between 0 and 1
data['length']=data['length']/data['length'].max()
data['width']=data['width']/data['width'].max()
data['height']=data['height']/data['height'].max()

#Binning-grouping groups
bins=np.linspace(min(data['price']),max(data['price']),4)
group_names=['Low','Medium','High']
data['price-binned']=pd.cut(data['price'],bins,labels=group_names,include_lowest=True)

print(data['price-binned'])
plt.hist(data['price-binned'])
plt.show()
```

```python
#ANOVA test to get the f-test and p-value
data_annova=data[['make','price']]
grouped_annova=data_annova.groupby(['make'])
annova_results_1=sp.stats.f_oneway(
                    grouped_annova.get_group('honda')['price'],
                    grouped_annova.get_group('subaru')['price'])
print(annova_results_1)

```
